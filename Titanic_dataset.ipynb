```python
#Q1 :What is the survival rate for males vs. females?
df.groupby('Sex')['Survived'].mean() * 100

#Answer : Females survived more.
```

```python
#Visualization
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#Q2:Did passengers in higher classes (1st class) have a higher survival rate compared to 
# lower classes (3rd class)?
df.groupby('Pclass')['Survived'].mean() * 100

#Answer : 1st class has more survival rate
```

```python
#Visualization
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Q3:Did passengers with more family members (siblings/spouses or parents/children) aboard 
# have a higher or lower chance of survival?
df['Family'] = df['SibSp'] + df['Parch'] + 1
df.groupby('Family')['Survived'].mean() * 100

#Answer :created a new feature named FamilySize by adding the number of siblings/spouses and parents/children on board, plus one for the passenger themselves.
# After calculating survival rates for different family sizes, we observed that passengers who were alone (FamilySize = 1) had a lower chance of survival.
# Passengers with small families (FamilySize between 2 and 4) had the highest survival rates, indicating that having some family support helped during evacuation.
# However, passengers with very large families (FamilySize > 5) had low survival rates, possibly due to difficulty coordinating escape as a group.
```

```python
#Q4:Did the port of embarkation (C, Q, S) influence the survival rate?
df.groupby('Embarked')['Survived'].mean() * 100

#Answer : Embarked C has more survival rate ie, passengers from Cherbourg(C) survived more
```

```python
#Visualization
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

```python
#Q5:Did passengers who paid higher fares have a better chance of survival?
#Diving Fare into 4 categories to compare in ranges.
df['Faregroups'] = pd.qcut(df['Fare'],4)
df.groupby('Faregroups')['Survived'].mean() * 100

#Answer : Passengers who paid higher fares had higher survival rates.
```

```python
#Dropping column cabin and name
df.drop(['Name','Cabin','Faregroups','Ticket'],axis=1,inplace=True)
df
# Cabin has too many missing values, so it does not add useful information.
# Name does not help in predicting survival (unless we extract titles like Mr, Mrs, etc.), so it can be removed for a clean model.
```

```python
#Encoding the categorical columns:
#Sex
#Embarked

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])
df['Embarked'] = encoder.fit_transform(df['Embarked'])
```

```python
x = df.drop(columns=['Survived'])
x
```

```python
y =df['Survived']
y
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#Machine learning model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
bayes = GaussianNB()
sv = SVC()
decesion = DecisionTreeClassifier(criterion='entropy')
random = RandomForestClassifier(n_estimators=100)
```

```python
lst = [knn,bayes,sv,decesion,random]
for i in lst:
    print("Model name is",i)
    i.fit(x_train,y_train)
    y_pred = i.predict(x_test)
    print(y_pred)
    print("*"*100)
    print(accuracy_score(y_test,y_pred))
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
```
